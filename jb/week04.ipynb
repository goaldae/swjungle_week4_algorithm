```python
# 2748 피보나치 수2 상향식, 하향식 두가지 방법이 있음

# 1 상향식
import sys

N = int(sys.stdin.readline())

arr = [0] * N

for i in range(N):
    if i == 0:
        arr[i] = 1
    elif i == 1:
        arr[i] = 1
    else:
        arr[i] = arr[i-1] + arr[i-2]

print(arr[-1])

# 2. 하향식
import sys
from collections import defaultdict

N = int(sys.stdin.readline())

memo = defaultdict(int)

print(memo)

def descend(n):
    if n <= 2:
        return 1
    if memo[n]:
        return memo[n]
    memo[n] = descend(n-1) + descend(n-2)
    return memo[n]

result = descend(N)

print(result)
```

```python
#(런타입에러 valueError) 12865 평범한 배낭 
import sys
N, K = map(int, sys.stdin.readline().split())

gene = dict()

for _ in range(N):
    a, b = map(int, sys.stdin.readline().strip().split())
    gene[a] = b

gene_key = sorted(gene.keys())

def optimize(k):
    if k in gene:
        return gene[k]

    bucket = []

    for i in gene_key:
        if k - i < gene_key[0] or k - i not in gene:
            continue
        elif gene[k-i] in gene:
            bucket.append(gene[k-i]) 
        else:
            value = optimize(k-i) + optimize(i)
            gene[k] = value
            bucket.append(value)

    return max(bucket)

result = optimize(K)
if result > 0:
    print(result)
else:
    print(0)
```

```python
# 11049 행렬곱셈 순서
import sys

N = int(sys.stdin.readline())
metrix = [list(map(int, sys.stdin.readline().split())) for _ in range(N)]

print(metrix)

graph = [[0 for _ in range(N)] for _ in range(N)]

print(graph)

for x in range(1, N):
    for i in range(N-x):
        j = i + x
        graph[i][j] = 2**31
        for k in range(i,j):
            graph[i][j] = min(graph[i][j], graph[i][k] + graph[k+1][j]+ metrix[i][0] * metrix[k][1] * metrix[j][1])


print(graph[0][-1])
```

```python
# 1931 (오답) 회의실배정
import sys

N = int(sys.stdin.readline())

data = []
for _ in range(N):
    a, b = map(int, sys.stdin.readline().split())
    data.append((a, b))

meetings = sorted(data, key=lambda x:(x[1], x[0]))

count = 1
while len(meetings) > 1:
    if meetings[0][0] == meetings[0][1]:
        meetings = meetings[1:]
    else:
        end_point = meetings[0][1]
        meetings = [x for x in meetings if x[0] >= end_point]
    count += 1


if count > 0:
    print(count)
else:
    print(0)

# 1931 회의실 배정


import sys

N = int(sys.stdin.readline())

data = []
for _ in range(N):
    a, b = map(int, sys.stdin.readline().split())
    data.append((a, b))

meetings = sorted(data, key=lambda x:(x[1], x[0]))

count = 1
end = meetings[0][1]

for i in range(1, len(meetings)):
    if meetings[i][0] >= end:
        count += 1
        end = meetings[i][1]

print(count)

    
```

```python
# 2253 (오답) 점프
# 이건 무식하게 dfs 다돌아서 초과나긴할텐데 왜 이거도 로컬에서 조차 답이 안나오나.. 해결해보삼

import sys
sys.setrecursionlimit(100000)

N, M = map(int, sys.stdin.readline().split())
trap = []

for _ in range(M):
    t = int(sys.stdin.readline())
    trap.append(t)


result = []

def pong_dang(point=1, speed=1, step = 0):
    if point in trap or speed <= 0:
        return -1
    
    if point == N:
        result.append(step)
        print(step)
        return step

    for j in range(-1, 2):
        point += speed
        speed += j
        step += 1
        pong_dang(point, speed, step)
        step -= 1
        speed -= j
        point -= speed

pong_dang()
print(result)
```

```python
# 11053 가장 긴 증가하는 부분 수열
import sys

N = int(sys.stdin.readline())

data = list(map(int, sys.stdin.readline().split()))

dp = [1 for _ in range(N)]

for i in range(len(data)):
    num_max = 0
    for j in range(0, i):
        if data[j] < data[i] and dp[j] > num_max:
            num_max = dp[j]
    dp[i] = num_max + 1

print(max(dp))
```

```python
# 2253 점프 (이해안감- 퍼온거임)
import sys
N, M = map(int, sys.stdin.readline().split())
dp = [[float('inf')] * (int((2 * N)** 0.5) + 2) for _ in range(N + 1)] 
dp[1][0] = 0

stone_set = set()
for _ in range(M):
    stone_set.add(int(sys.stdin.readline()))
for i in range(2, N + 1):
    if i in stone_set:
        continue
    for j in range(1, int((2 * i) ** 0.5) + 1):
        dp[i][j] = min(dp[i - j][j - 1], dp[i - j][j], dp[i - j][j + 1]) + 1

if min(dp[N]) == float('inf'):
    print(-1)
else:
    print(min(dp[N]))
```
